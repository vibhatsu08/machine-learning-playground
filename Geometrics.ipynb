```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
people = np.arange(1, 11)
total_cost = np.ones_like(people) * 40.0
people, total_cost
```

```python
axes = plt.gca()
axes.set_xlabel("# People")
axes.set_ylabel("# Total Cost")
axes.plot(people, total_cost)
```

```python
people = np.arange(1, 11)
total_cost = 80.0 * people + 40.0
people, total_cost
```

```python
# making a table of the total_cost values array.
display = pd.DataFrame(
    {
        "total_cost": total_cost.astype(int)
    },
    index = people
).T
display
```

```python
axes = plt.gca()
axes.plot(people, total_cost, "bo")
axes.set_xlabel("# People")
axes.set_ylabel("# Total Cost")
```

```python
xs = np.linspace(-3, 3, 100)
m, b = 1.5, -3
axes = plt.gca()
ys = m*xs + b
axes.plot(xs, ys)
axes.plot(0, -3, "ro")
axes.plot(2, 0, "ro")
# y = mx + b, with m = 0, gives y = b
y = 0*xs + b
axes.plot(xs, ys, "y")
```
