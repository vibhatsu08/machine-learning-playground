```python
import numpy as np
import pandas as pd
from sklearn import (
    datasets, metrics, model_selection as skms, dummy
)
```

```python
dataset = datasets.load_iris()
```

```python
(train_ftrs, test_ftrs, train_trgt, test_trgt) = skms.train_test_split(dataset.data, dataset.target, test_size=0.25)
train_ftrs, test_ftrs, train_trgt, test_trgt
```

```python
baseline = dummy.DummyClassifier(strategy="most_frequent")
fit = baseline.fit(train_ftrs, train_trgt)
predict = fit.predict(test_ftrs)
accuracy = metrics.accuracy_score(predict, test_trgt)
print(accuracy)
```

```python
strategies = [{
    "strategy": "constant",
    "constant": 0
},
     "uniform", "stratified", "prior", "most_frequent"
]

scores = []
for strat in strategies:
    if isinstance(strat, dict):
        baseline = dummy.DummyClassifier(**strat)
    else:
        baseline = dummy.DummyClassifier(strategy=strat)
    fit = baseline.fit(train_ftrs, train_trgt)
    predict = fit.predict(test_ftrs)
    score = metrics.accuracy_score(predict, test_trgt)
    scores.append(score)
df = pd.DataFrame({"baseline_score": scores}, index=strategies)
df
```
