```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import (
    datasets, metrics, model_selection as skms, naive_bayes, neighbors
)
```

```python
dataset = datasets.load_iris()
```

```python
models = {
    "gnb": naive_bayes.GaussianNB(),
    "5-NN": neighbors.KNeighborsClassifier(n_neighbors=5)
}
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for name, model in models.items():
    cv_scores = skms.cross_val_score(model, dataset.data, dataset.target, cv=10, scoring="accuracy", n_jobs=-1)
    model_label = f"{name} : {cv_scores.mean():.3f}"
    ax.plot(cv_scores, "-o", label=model_label)
ax.set_ylim(0.1, 1.1)
ax.set_xlabel("Fold")
ax.set_ylabel("Accuracy")
ax.legend(ncol=2, loc="lower left")
```
