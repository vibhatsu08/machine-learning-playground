```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import (
    datasets, metrics, model_selection as skms, neighbors, naive_bayes
)
```

```python
dataset = datasets.load_iris()
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors=5)
```

```python
loo = skms.LeaveOneOut()
```

```python
evaluations = [loo, 2, 3, 5, 10]
labels = ["LeaveOneOut", "2-Fold", "3-Fold", "5-Fold", "10-Fold"]
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for label, eval in zip(labels, evaluations):
    cv_score = skms.cross_val_score(model, dataset.data, dataset.target, cv=eval,scoring="accuracy", n_jobs=-1)
    score_label = "CV evaluation - {} : {:.3f}".format(label, cv_score.mean())
    ax.plot(cv_score, "o-", label=score_label)
ax.set_xlabel("Evaluations (loo, 2, 3, 5, 10)")
ax.set_ylabel("Accuracy")
ax.legend(loc="best")
ax.set_ylim(0.0, 1.1)
ax.set_xlim(-1, 10)
```

```python
model = naive_bayes.GaussianNB()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for label, eval in zip(labels, evaluations):
    cv_score = skms.cross_val_score(model, dataset.data, dataset.target, cv=eval,scoring="accuracy", n_jobs=-1)
    score_label = "CV evaluation - {} : {:.3f}".format(label, cv_score.mean())
    ax.plot(cv_score, "o-", label=score_label)
ax.set_xlabel("Evaluations (loo, 2, 3, 5, 10)")
ax.set_ylabel("Accuracy")
ax.legend(loc="best")
ax.set_ylim(0.0, 1.1)
ax.set_xlim(-1, 10)
```
