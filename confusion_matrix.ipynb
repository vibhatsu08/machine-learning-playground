```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import (
    datasets, metrics, model_selection as skms, neighbors
)
```

```python
dataset = datasets.load_iris()
```

```python
(train_ftrs, test_ftrs, train_trgt, test_trgt) = skms.train_test_split(dataset.data, dataset.target, test_size=0.25)
```

```python
predict = (neighbors.KNeighborsClassifier()
                    .fit(train_ftrs, train_trgt)
                    .predict(test_ftrs))
score = metrics.accuracy_score(test_trgt, predict)
score
```

```python
# creating the confusion matrix.
cm = metrics.confusion_matrix(test_trgt, predict)
print(cm)
```

```python
# plotting the confusion matrix.
fig, ax = plt.subplots(figsize=(10, 6))
cm = metrics.confusion_matrix(test_trgt, predict)
ax = sns.heatmap(cm, annot=True, square=True,
                xticklabels=dataset.target_names,
                yticklabels=dataset.target_names)
ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
```
