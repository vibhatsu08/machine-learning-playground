```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

```python
dataset = datasets.load_iris()
dataframe = pd.DataFrame(dataset.data, columns=dataset.feature_names)
dataframe['target'] = dataset.target
display(pd.concat([dataframe.head(3), dataframe.tail(3)]))
dataframe
```

```python
sns.pairplot(dataframe, hue='target', height=1.5)
```

```python
print(f"target: {dataset.target_names}")
print(f"target: {dataset.target_names[0]}")
# from the visualisation it is understood that the setosa, target 0, tends to stand apart from the other two in the dataset in terms of similarity.
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataset_train_ftrs, dataset_test_ftrs, dataset_train_trgt, dataset_test_trgt = train_test_split(dataset.data, dataset.target, test_size=0.25)
dataset_train_ftrs, dataset_test_ftrs, dataset_train_trgt, dataset_test_trgt
```

```python
dataset_train_ftrs.shape, dataset_test_ftrs.shape
```

```python
# training a k-nn model to fit, predict and evaluate the accuracy of the predictions.
from sklearn.neighbors import KNeighborsClassifier
# importing metrics to be able to measure the accuracy of the model
from sklearn import metrics
```

```python
# creating the 3-nn model. default value of n_neighbors is 5.
knn = KNeighborsClassifier(n_neighbors=3)

# fitting the above model on the training data.
fit = knn.fit(dataset_train_ftrs, dataset_train_trgt)

# using the model to predict on the test data.
preds = fit.predict(dataset_test_ftrs)

# evaluating the predictions accuracy.
accuracy = metrics.accuracy_score(dataset_test_trgt, preds)
accuracy
```

```python
# training a Naive Bayes model to fit, predict and evaluate the accuracy of the predictions.
from sklearn.naive_bayes import GaussianNB
```

```python
# creating the NB model.
nb = GaussianNB()

# fitting the above model on the training features and training targets.
fit = nb.fit(dataset_train_ftrs, dataset_train_trgt)

# using the model to predict on the test features
preds = fit.predict(dataset_test_ftrs)

# evaluating the predictions accuracy
accuracy = metrics.accuracy_score(dataset_test_trgt, preds)
accuracy
```
